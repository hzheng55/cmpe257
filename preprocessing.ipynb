```
#obtain only articles in 2020, 2019, and 2018. place into articles_new. save articles_new into csv file.
import numpy as np

def find_articles_by_year(year):
  s = str(year)
  find_year = metadata["id"].str.find(s)
  found = find_year[find_year.values == 1]
  found_new_index = found.reset_index()
  return_find= metadata.iloc[found_new_index["index"]]
  return return_find.reset_index()

def find_articles_by_id(file_extra_location):
  f = open(file_extra_location, "r")
  return_find = pd.DataFrame()
  file = f.readlines()
  for i in file:
    id = "\""+i[0:10]+"\""
    m=metadata[metadata["id"]==id]
    return_find = return_find.append(m)
  return return_find

file_extra = os.path.join(path, "extra articles.txt")
extra_articles = find_articles_by_id(file_extra)

articles_in_2020 = find_articles_by_year(2003)
articles_in_2019 = find_articles_by_year(1903)
articles_in_2018 = find_articles_by_year(1803)

articles_in_2020 = articles_in_2020.head(30)
articles_in_2019 = articles_in_2019.head(30)
articles_in_2018 = articles_in_2018.head(30)

articles = [articles_in_2020, articles_in_2019, articles_in_2018, extra_articles]
articles_new = pd.concat(articles)
articles_new = articles_new.drop(columns = "index")
articles_new.head()
articles_new = articles_new.reset_index()

articles_new.to_csv("/content/drive/My Drive/last-3-years.csv")
```
